```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline([
     ('scale',StandardScaler() )
])
```

```python
num_attrs=list(X_train_num.columns)
print(num_attrs)
nominal_attrs=list(X_train_cat_main.columns)
print(nominal_attrs)
```

```python
pipeline=ColumnTransformer([
    ('numerical',num_pipeline,num_attrs),
    ('nominal',OneHotEncoder(),nominal_attrs)
])
```

```python
data=pipeline.fit_transform(X_train)
data[0:5]
```

```python
num_cols=num_attrs
cat_cols = list(pipeline.named_transformers_['nominal'].categories_[:])    
columns=num_cols+cat_cols
print(columns)
```

## Error calculation

```python
from sklearn.metrics import mean_squared_error
```

```python
def rmse(y_actual,y_predicted):
    return np.sqrt((y_actual-y_predicted)**2)
    
```

```python
e=rmse(actual,predicted)
```

```python
e
```

```python
error=pd.DataFrame({'Error': e.values})
```
